```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df = pd.read_csv('SampleSuperstore.csv', encoding='ISO-8859-1')
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Calculate total profit by Category
profit_by_Category = df.groupby("Category")["Profit"].sum()

# Calculate total profit
total_profit = profit_by_Category.sum()

# Calculate percentage of total profit
percentage_profit = (profit_by_Category / total_profit) * 100

# Plotting the bar graph
plt.figure(figsize=(8, 6))
bars = plt.bar(percentage_profit.index, percentage_profit, linewidth=1)

# Adding annotations to the bars
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # Center of the bar
        bar.get_height() + 0.5,            # Slightly above the bar
        f"{bar.get_height():.2f}%",       # Display value with 2 decimal places
        ha='center', va='bottom', fontsize=12
    )

# Adding labels and title
plt.title("Percentage of Total Profit by Category", fontsize=14)
plt.xlabel("Category", fontsize=12)
plt.ylabel("Percentage of Total Profit (%)", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Step 1: Filter the data for Accessories subcategory
accessories_data = df[df['Sub-Category'] == 'Accessories']

# Step 2: Calculate profit per unit for each product
accessories_data['Profit per Unit'] = accessories_data['Profit'] / accessories_data['Quantity']

# Step 3: Sort by 'Profit per Unit' in descending order
sorted_accessories = accessories_data.sort_values(by='Profit per Unit', ascending=False)

# Step 4: Select the top products that total 757 additional quantity
sorted_accessories['Cumulative Quantity'] = sorted_accessories['Quantity'].cumsum()
additional_products = sorted_accessories[sorted_accessories['Cumulative Quantity'] <= 757]

# Step 5: Calculate the total profit from the additional products
total_profit_added = additional_products['Profit'].sum()

# Step 6: Calculate the percentage contribution to total profit
total_profit_before = df['Profit'].sum()  # Total profit in the original dataset
percentage_increase = (total_profit_added / total_profit_before) * 100

# Step 7: Display the results
print(f"Total Profit from Additional 757 Quantities: ${total_profit_added:.2f}")
print(f"Percentage Increase in Total Profit: {percentage_increase:.2f}%")
```
